```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName("Distributed Data Processing Activity") \
.getOrCreate()

spark
```

```python
data = [
    ("Laptop", "Electronics", 1200),
    ("Phone", "Electronics", 800),
    ("Table", "Furniture", 300),
    ("Chair", "Furniture", 150),
    ("Headphones", "Electronics", 200),
    ("Sofa", "Furniture", 700),
    ("Monitor", "Electronics", 400),
    ("Bed", "Furniture", 900)
]

columns = ["product", "category", "price"]

df = spark.createDataFrame(data, columns)

df.show()
```

```python
from pyspark.sql.functions import col

filtered_df = df.filter(col("price") > 500)

filtered_df.show()
```

```python
grouped_df = df.groupBy("category").count()

grouped_df.show()
```

```python
sorted_df = grouped_df.orderBy("count", ascending=False)

#sorted_df.show()
```

# PART III: PERFORMANCE AND SCALABILITY DISCUSSION
### How does Spark divide data?

Spark divides datasets into smaller parts called partitions. Each partition is processed independently by different CPU cores, allowing tasks to run simultaneously instead of sequentially.

### What happens if dataset size increases?

If the dataset becomes larger, Spark automatically creates more partitions. The system can distribute the workload across more computing resources, allowing the program to scale efficiently.

### Why is this better than standard Python processing?

Standard Python typically processes tasks sequentially. This means operations run one after another, which becomes slow for large datasets. Spark improves performance by using parallel processing and distributed computing, making it much faster for big data tasks.

### Why is Google Colab considered cloud computing?

Google Colab is considered cloud computing because the code runs on Google's remote servers instead of the user's local computer. Users can access computing resources like CPU, RAM, and storage through the internet.

### Which cloud service model does this resemble?

Google Colab resembles Platform as a Service (PaaS) because it provides a ready-to-use development environment where users can run code without managing the underlying infrastructure.

### How does cloud infrastructure help distributed systems?

Cloud infrastructure provides scalable computing resources, multiple servers, and flexible storage. This allows distributed systems like Spark to process large datasets efficiently and scale when needed.
